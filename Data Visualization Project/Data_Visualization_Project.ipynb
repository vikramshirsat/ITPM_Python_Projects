# 2. Visualization Graph using Pandas, Matplotlib, Seaborn

## 1. Line Plots

```python
plt.figure(figsize=(8, 6))  
plt.plot(df.head(10)['player_name'], df.head(10)['total_runs'], marker="o", label='Player vs total runs', linestyle="-")
plt.title('Player vs total runs')
plt.xlabel('player_name')
plt.ylabel('total_runs')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True, linestyle='--',color = "black" ,alpha=0.8)
plt.gca().set_facecolor('lightgray') 

plt.xticks(rotation=45, ha='right')  
plt.tight_layout()  
plt.show()
```

## 2. Bar Chart

```python
x = df["team"]
y = df["matches_won"]

plt.figure(figsize=(12, 6))  
plt.bar(x, y, color="blue")
plt.title('Cricket Team with Matches Won')
plt.xlabel('Teams')
plt.ylabel('Matches Won')

plt.xticks(rotation=45, ha='right')  
plt.tight_layout()  
plt.show()
```

## 3. Histograms

```python
plt.figure(figsize=(10,6))
plt.hist(df['total_runs'], bins=10, color='orange', edgecolor='black')
plt.title("Distribution of Total Runs")
plt.xlabel("Total Runs")
plt.ylabel("Number of Players")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['total_matches_played'], bins=20, kde=False, color='skyblue')
plt.title('Distribution of Total Matches Played', fontsize=16)
plt.xlabel('Total Matches Played', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

## 4. Scatter Plots

```python
import matplotlib.pyplot as plt

x = df.head(10)['team']
y = df.head(10)['percentage']

plt.figure(figsize=(12,6))  
plt.scatter(x, y, color='red')
plt.title('Team vs Percentage')
plt.xlabel('Teams')
plt.ylabel('Percentage')

plt.xticks(rotation=45, ha='right') 
plt.yticks(rotation=45, ha='right')
plt.gca().set_facecolor('skyblue') 
plt.tight_layout()  
plt.show()
```

## 5. Heatmap

```python
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap='plasma', fmt='.1f', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

## 6. Pair Plots

```python
sns.pairplot(df, hue='team', diag_kind='kde', palette='husl')
plt.show()
```

## 7. boxplot

```python
plt.figure(figsize=(15, 6))  
sns.boxplot(x='team', y='matches_won', data=df, hue='team', palette='Set2')
plt.title("Box Plot of matches_won by team")
plt.xlabel("Team")
plt.ylabel("Matches Won")
plt.xticks(rotation=45, ha='right')  
plt.show()
```

## 8. countplot

```python
# Countplot for 'role'
plt.figure(figsize=(10, 6))
sns.countplot(x='role', data=df, hue='role', palette='Set2')
plt.title('Count of Players by Role')
plt.xlabel('Role')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Countplot for 'team'
plt.figure(figsize=(12, 6))
sns.countplot(x='team', data=df, hue='team', palette='Set3')
plt.title('Count of Players by Team')
plt.xlabel('Team')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## 9. pie chart

```python
team_counts = df['team'].value_counts()

plt.figure(figsize=(10, 8))
plt.pie(
    team_counts,
    labels=team_counts.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Players by Team')
plt.tight_layout()
plt.show()
```

## 10. bar plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x='team', y='matches_won', data=df, estimator=sum, hue='team', palette='viridis')

plt.title('Total Matches Won by Each Team')
plt.xlabel('Teams')
plt.ylabel('Total Matches Won')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
