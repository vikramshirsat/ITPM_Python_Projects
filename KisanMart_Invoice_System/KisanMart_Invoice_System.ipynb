## KisanMart Invoice System

```python
product_category = [
    "Seeds",
    "Fertilizers",
    "Pesticides / Insecticides",
    "Irrigation & Equipment",
    "Animal Feed & Supplements"
]

product_lists = {
    "Seeds": [
        {"Hybrid Tomato Seeds": 120},
        {"Onion Seeds": 95},
        {"Maize Seeds": 210},
        {"Paddy Seeds": 85},
        {"Chilli Seeds": 150}
    ],
    "Fertilizers": [
        {"Urea": 1050},
        {"DAP": 1400},
        {"NPK 20:20:0": 1350},
        {"Potash": 1250},
        {"Zinc Sulphate": 85}
    ],
    "Pesticides / Insecticides": [
        {"Rogor": 680},
        {"Bavistin": 85},
        {"Malathion": 520},
        {"Neem Oil": 450},
        {"Glyphosate": 720}
    ],
    "Irrigation & Equipment": [
        {"Battery Sprayer": 3500},
        {"Hand Sprayer": 850},
        {"Drip Kit": 2500},
        {"Water Pipe": 35},
        {"Motor Pump": 7800}
    ],
    "Animal Feed & Supplements": [
        {"Poultry Feed": 1650},
        {"Cattle Feed": 1550},
        {"Mineral Mixture": 85},
        {"Salt Brick": 35},
        {"Vitamin Supplement": 420}
    ]
}

cart = []

def menu():
    print("\n=== KisanMart Main Menu ===")
    for i in range(len(product_category)):
        print(f"{i + 1}. {product_category[i]}")
    print("6. Exit")

def show_products(category):
    print(f"\n--- {category} ---")
    for i in range(len(product_lists[category])):
        for name, price in product_lists[category][i].items():
            print(f"{i + 1}. {name} - ₹{price}")

def select_product(category):
    show_products(category)
    choice = int(input("Select product number (1–5): "))
    qty = int(input("Enter quantity: "))
    
    item = list(product_lists[category][choice - 1].items())[0]
    name, price = item
    total = price * qty

    cart.append([name, qty, price, total])
    print(f"Added {qty} × {name} = ₹{total}")

def show_cart():
    print("\n=== Your Cart ===")
    total_amount = 0
    for item in cart:
        print(f"{item[0]} (x{item[1]}) - ₹{item[3]}")
        total_amount += item[3]
    print(f"\nTotal Amount: ₹{total_amount}")
    return total_amount

def payment(total):
    while True:  
        print("\nChoose Payment Method:")
        print("1. Cash")
        print("2. Credit/Debit Card")
        print("3. UPI")
        option = int(input("Enter choice for payment (1–3): "))

        if option == 1:
            print("\nPayment Successful by Cash!")
            break
        elif option == 2:
            print("\nPayment Successful by Card!")
            break
        elif option == 3:
            print("\nPayment Successful by UPI!")
            break
        else:
            print("\nInvalid Payment Option! Please try again.\n")

    print("\nThank you for shopping with us!")

while True:
    menu()
    choice = int(input("\nEnter category number (1–6): "))

    if choice == 6:
        total = show_cart()
        if total > 0:
            payment(total)
        else:
            print("\nNo items in cart.")
        break
    elif 1 <= choice <= 5:
        category = product_category[choice - 1]
        select_product(category)
    else:
        print("Invalid choice. Please try again.")
```
